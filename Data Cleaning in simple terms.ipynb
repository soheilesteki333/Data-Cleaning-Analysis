We start the **Fix Date Type**
=====

```python
clean_df = clean_df.copy()
clean_df ["Transaction Date"] = pd.to_datetime(clean_df["Transaction Date"])
clean_df
```

```python
clean_df.info()
```

Engineereing time-based features
==

```python
clean_df["month"] = clean_df["Transaction Date"].dt.month
clean_df["year"] = clean_df["Transaction Date"].dt.year
clean_df
```

```python
clean_df.groupby("month")["Total Spent"].sum().sort_values(ascending=False)
```

```python
clean_df.groupby("Category")["Total Spent"].sum().sort_values(ascending=False)
```

```python
clean_df.to_csv("cleaned_retail_sales.csv", index=False)
```
